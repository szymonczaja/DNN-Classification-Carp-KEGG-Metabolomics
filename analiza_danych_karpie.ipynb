```python
from scipy.stats import shapiro

ko_cols = [col for col in df.columns if col.startswith('ko')]

normalne_count = 0
nienormalne_count = 0

for col in ko_cols:
    data = df[col].dropna()
    stat, p = shapiro(data)

    if p > 0.05:
        normalne_count += 1
    else:
        nienormalne_count += 1

print("Wyniki testÃ³w normalnoÅ›ci")
print(f"Normalne:     {normalne_count}")
print(f"Nienormalne:  {nienormalne_count}")

```

NormalnoÅ›Ä‡ rozkÅ‚adu oceniono przy uÅ¼yciu testu Shapiroâ€“Wilka. Wyniki pokazujÄ…, Å¼e zdecydowana wiÄ™kszoÅ›Ä‡ cech (113 z 155) nie ma rozkÅ‚adu normalnego.
W zwiÄ…zku zastosowana zostaÅ‚a nieparametryczna metoda korelacji Spearmana, ktÃ³ra nie wymaga normalnoÅ›ci danych i jest odporna na wartoÅ›ci odstajÄ…ce.

```python
import seaborn as sns
import matplotlib.pyplot as plt

ko_cols = [col for col in df.columns if col.startswith('ko')]

corr_with_target = df[ko_cols].corrwith(df['pond_5'], method='spearman')
corr_with_target = corr_with_target.sort_values(ascending=False)

corr_matrix = df[ko_cols].corr(method='spearman')

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title("Macierz korelacji Spearmana")
plt.show()
```

```python
from scipy.stats import spearmanr

data_ko = ko
to_drop = []

for i in range(len(ko_cols)):
    for j in range(i+1, len(ko_cols)):
        ko1 = ko_cols[i]
        ko2 = ko_cols[j]
        r, p = spearmanr(data_ko[ko1], data_ko[ko2])

        if abs(r) > 0.8:
            if data_ko[ko1].var() > data_ko[ko2].var():
                if ko2 not in to_drop:
                    to_drop.append(ko2)
            else:
                if ko1 not in to_drop:
                    to_drop.append(ko1)

to_drop = list(set(to_drop))
reduced = data_ko.drop(columns=to_drop)

print("po redukcji:", reduced.shape[1])
```

```python
corr_matrix_reduced = reduced.corr(method='spearman')
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix_reduced, cmap='coolwarm', center=0)
plt.title("Macierz korelacji Spearmana")
plt.show()
```

```python
def create_dnn_pca():
  model_pca = tf.keras.Sequential([
    tf.keras.layers.InputLayer(shape=[17]),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.6),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
  ])
  model_pca.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
  return model_pca

def create_dnn_corr():
  model_corr = tf.keras.Sequential([
    tf.keras.layers.InputLayer(shape=[35]),
    tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(5, activation='softmax')

  ])
  model_corr.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
  return model_corr

def create_dnn_all():
  model_all = tf.keras.Sequential([
    tf.keras.layers.InputLayer(shape=[155]),
    tf.keras.layers.Dense(128, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=l2(0.002)),
    tf.keras.layers.Dense(5, activation='softmax')
  ])
  model_all.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
  return model_all
```

```python
def run_and_collect_score(n=10):
  models_to_test = [
      ('DNN_PCA', create_dnn_pca, X_train_pca_array, X_test_pca_array, y_train_pca_ohe, y_test_pca_ohe, 'Keras'),
      ('DNN_Korelacje', create_dnn_corr, X_train_array, X_test_array, y_train_one_hot, y_test_one_hot, 'Keras'),
      ('DNN_AllKeggs', create_dnn_all, X_train2_array, X_test2_array, y_train_one_hot2, y_test_one_hot2, 'Keras'),
      ('RandomForest_Korelacje', rf_clf_corr, X_train, X_test, y_train, y_test, 'Scikit-Learn'),
      ('LineraSVC_PCA', svc_clf, X_train_pca, X_test_pca, y_train, y_test, 'Scikit-Learn')
  ]
  all_results = []
  for run in range(n):
    print(f'--Uruchomienie {run+1}/{n}')
    for name, model_def, X_tr, X_te, y_tr, y_te, model_type in models_to_test:
      if model_type == 'Keras':
        tf.keras.backend.clear_session()
        model = model_def()
        history = model.fit(X_tr, y_tr,
                            epochs=300,
                            validation_data=(X_te, y_te),
                            callbacks=[EarlyStopping(patience=40, restore_best_weights=True)],
                            verbose=0)
        train_loss, train_accuracy = model.evaluate(X_tr, y_tr, verbose=0)
        test_loss, test_accuracy = model.evaluate(X_te, y_te, verbose=0)
      else:
        model = model_def
        model.fit(X_tr, y_tr)
        y_pred_train = model.predict(X_tr)
        y_pred_test = model.predict(X_te)
        train_accuracy = accuracy_score(y_tr, y_pred_train)
        test_accuracy = accuracy_score(y_te, y_pred_test)

      all_results.append({
          'Model': name,
          'AccuracyType': 'Train Accuracy',
          'Accuracy' : train_accuracy
      })
      all_results.append({
          'Model': name,
          'AccuracyType': 'Test Accuracy',
          'Accuracy' : test_accuracy
      })
  return pd.DataFrame(all_results)
```
